```python
import torch
import torch.nn as nn
from torch.utils.data import TensorDataset, DataLoader 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Classical

Importing and converting data.

```python
df = pd.read_csv('brazil_covid19_macro.csv')
cases = df['cases'].values

cases_day = np.array([cases[i + 1] - cases[i] for i in range(cases.shape[0] - 1)], dtype='float32')
days = np.array([i for i in range(cases.shape[0] - 1)], dtype='float32')

cases_day /= cases_day.max()
days /= days.max()
```

Plotting.

```python
plt.figure(figsize=(12, 5))
plt.scatter(days, cases_day, marker='s')
plt.title('COVID-19 cases per day in Brazil')
plt.ylabel('Normalised cases')
plt.xlabel('Normalised days')
plt.show()

kernel = np.ones(15) / 15
mean_cases_day = np.convolve(cases_day, kernel, mode='valid')
mean_days = days[:mean_cases_day.shape[0]]
mean_days /= mean_days.max()

plt.figure(figsize=(12, 5))
plt.scatter(mean_days, mean_cases_day, marker='s')
plt.title('COVID-19 cases per day in Brazil')
plt.ylabel('Normalised cases')
plt.xlabel('Normalised days')
plt.show()
```

Creating sequences.

```python
def create_sequences(data, time, seq_length):
    X, y, time_labels = [], [], []
    for i in range(len(data) - seq_length): 
        X.append(data[i : i + seq_length])  
        y.append(data[i + seq_length])      
        time_labels.append(time[i + seq_length])  
    return np.array(X), np.array(y), np.array(time_labels)

seq_length = 17
X, y, prediction_days = create_sequences(mean_cases_day, days, seq_length)

X = torch.FloatTensor(X).unsqueeze(2)  
y = torch.FloatTensor(y).unsqueeze(1) 

train_size = int(0.8 * len(X))
val_size = int(0.1 * len(X))

X_train, y_train, days_train = X[:train_size], y[:train_size], prediction_days[:train_size]
X_val, y_val, days_val = X[train_size:train_size+val_size], y[train_size:train_size+val_size], prediction_days[train_size:train_size+val_size]
X_test, y_test, days_test = X[train_size+val_size:], y[train_size+val_size:], prediction_days[train_size+val_size:]

train_dataset = TensorDataset(X_train, y_train)
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=False)
```

Instantianting RNN.

```python
class NN(nn.Module):
    def __init__(self, h=64, activation=nn.Sigmoid()):
        super().__init__()
        self.rnn = nn.RNN(
            input_size=1,      
            hidden_size=h,    
            batch_first=True   
        )
        self.fc = nn.Linear(h, 1)
        self.activation = activation
    
    def forward(self, x):
        out, _ = self.rnn(x)         
        return self.activation(self.fc(out[:, -1, :]))

model = NN(
    h=110,
    activation=nn.Identity(),
)

params_rnn = 0
for p in list(model.parameters()):
    params_rnn += p.flatten().shape[0]
    
print(f'Number of parameters: {params_rnn}')
```

Training the model.

```python
try:
    best_model_path = 'best_models/rnn.pth'  
    checkpoint = torch.load(best_model_path, weights_only=True)
    model.load_state_dict(checkpoint['model_state_dict'])
    print('Loaded.')

except:
    criterion = nn.MSELoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=0.0013534016965942425, weight_decay=1.0097368709223582e-05)
    
    epochs = 1000
    best_val_loss = float('inf')
    best_model_path = 'best_models/rnn.pth'
    
    for epoch in range(epochs):
        model.train()
        epoch_loss_train = 0.0
        for X_batch, y_batch in train_loader:
            optimizer.zero_grad()
            outputs = model(X_batch)
            loss = criterion(outputs, y_batch)
            loss.backward()
            optimizer.step()
            epoch_loss_train += loss.item() * X_batch.size(0) 
    
        epoch_loss_train /= len(train_loader.dataset)
        
        model.eval()
        with torch.no_grad():
            outputs = model(X_val)
            loss_val = criterion(outputs, y_val)
    
        if loss_val < best_val_loss:
                best_val_loss = loss_val
                torch.save({
                    'model_state_dict': model.state_dict(),
                }, best_model_path)
        
        if epoch % 50 == 0:
            print(f'Epoch {epoch} | Train Loss: {epoch_loss_train:.5f} | Validation Loss: {loss_val.item():.5f}')
    
    checkpoint = torch.load(best_model_path, weights_only=True)
    model.load_state_dict(checkpoint['model_state_dict'])
```

Plotting training data predictions.

```python
model.eval()
with torch.no_grad():
    train_predictions_rnn = model(X_train).numpy().flatten()

plt.figure(figsize=(15, 4))

plt.subplot(1, 2, 1)
plt.scatter(days_train, y_train.numpy(), label='True Values', alpha=0.5)
plt.scatter(days_train, train_predictions_rnn, label='Predictions')
plt.legend()
plt.title("Training Fit")

plt.subplot(1, 2, 2)
plt.scatter(days_train, y_train.numpy().flatten() - train_predictions_rnn)
plt.axhline(0, c='red', linestyle='--')
plt.title('Error')

plt.tight_layout()  
plt.show()
```

```python
model.eval()
with torch.no_grad():
    val_predictions_rnn = model(X_val).numpy().flatten()

plt.figure(figsize=(15, 4))

plt.subplot(1, 2, 1)
plt.scatter(days_val, y_val.numpy(), label='True Values', alpha=0.5)
plt.scatter(days_val, val_predictions_rnn, label='Predictions')
plt.legend()
plt.title("Validation Data")

plt.subplot(1, 2, 2)
plt.scatter(days_val, y_val.numpy().flatten() - val_predictions_rnn)
plt.axhline(0, c='red', linestyle='--')
plt.title('Validation Error')

plt.tight_layout()
plt.show()
```

```python
model.eval()
with torch.no_grad():
    test_predictions_rnn = model(X_test).numpy().flatten()

plt.figure(figsize=(15, 4))

plt.subplot(1, 2, 1)
plt.scatter(days_test, y_test.numpy(), label='True Values', alpha=0.5)
plt.scatter(days_test, test_predictions_rnn, label='Predictions')
plt.legend()
plt.title("Test Data")

plt.subplot(1, 2, 2)
plt.scatter(days_test, y_test.numpy().flatten() - test_predictions_rnn)
plt.axhline(0, c='red', linestyle='--')
plt.title('Test Error')

plt.tight_layout()
plt.show()
```

## Fuzzy

```python
from ANFISpy import RANFIS
```

Instantiating model.

```python
n_vars = 1
mf_names = [['VL', 'L', 'ML', 'M', 'MH', 'H', 'VH']]

variables = {
    'inputs': {
        'n_sets': [7],
        'uod': n_vars * [(0, 1)],
        'var_names': ['t'],
        'mf_names': n_vars * mf_names,
    },
    'output': {
        'var_names': 'out',
        'n_classes': 1,
    },
}

ranfis = RANFIS(variables, 'gaussian', output_activation=nn.Identity(), and_operator=torch.prod)

params_ranfis = 0
for p in list(ranfis.parameters()):
    params_ranfis += p.flatten().shape[0]
    
print(f'Number of parameters: {params_ranfis}')
```

Separating data.

```python
def create_sequences(data, time, seq_length):
    X, y, time_labels = [], [], []
    for i in range(len(data) - seq_length): 
        X.append(data[i : i + seq_length])  
        y.append(data[i + seq_length])      
        time_labels.append(time[i + seq_length])  
    return np.array(X), np.array(y), np.array(time_labels)

seq_length = 17 
X, y, prediction_days = create_sequences(mean_cases_day, days, seq_length)

X = torch.FloatTensor(X).unsqueeze(2)  
y = torch.FloatTensor(y).unsqueeze(1) 

train_size = int(0.8 * len(X))
val_size = int(0.1 * len(X))

X_train, y_train, days_train = X[:train_size], y[:train_size], prediction_days[:train_size]
X_val, y_val, days_val = X[train_size:train_size+val_size], y[train_size:train_size+val_size], prediction_days[train_size:train_size+val_size]
X_test, y_test, days_test = X[train_size+val_size:], y[train_size+val_size:], prediction_days[train_size+val_size:]

train_dataset = TensorDataset(X_train, y_train)
train_loader = DataLoader(train_dataset, batch_size=16, shuffle=False)
```

Training.

```python
try:
    best_model_path = 'best_models/ranfis.pth'  
    checkpoint = torch.load(best_model_path, weights_only=True)
    ranfis.load_state_dict(checkpoint['model_state_dict'])
    print('Loaded.')

except:
    criterion = nn.MSELoss()
    optimizer = torch.optim.Adam(ranfis.parameters(), lr=0.009138, weight_decay=0.000036)
    
    epochs = 1000
    best_val_loss = float('inf')
    best_model_path = 'best_models/ranfis.pth'
    
    for epoch in range(epochs):
        ranfis.train()
        epoch_loss_train = 0.0
        for X_batch, y_batch in train_loader:
            optimizer.zero_grad()
            outputs = ranfis(X_batch)[0]
            loss = criterion(outputs, y_batch)
            loss.backward()
            optimizer.step()
            epoch_loss_train += loss.item() * X_batch.size(0) 
    
        epoch_loss_train /= len(train_loader.dataset)
        
        ranfis.eval()
        with torch.no_grad():
            outputs = ranfis(X_val)[0]
            loss_val = criterion(outputs, y_val)
    
        if loss_val < best_val_loss:
            best_val_loss = loss_val
            torch.save({
                'model_state_dict': ranfis.state_dict(),
            }, best_model_path)
        
        if epoch % 50 == 0:
            print(f'Epoch {epoch} | Train Loss: {epoch_loss_train:.5f} | Validation Loss: {loss_val.item():.5f}')
    
    checkpoint = torch.load(best_model_path, weights_only=True)
    ranfis.load_state_dict(checkpoint['model_state_dict'])
```

Plotting.

```python
ranfis.eval()
with torch.no_grad():
    train_predictions_ranfis = ranfis(X_train)[0].numpy().flatten()

plt.figure(figsize=(15, 4))

plt.subplot(1, 2, 1)
plt.scatter(days_train, y_train.numpy(), label='True Values', alpha=0.5)
plt.scatter(days_train, train_predictions_ranfis, label='Predictions')
plt.legend()
plt.title("Training Fit")

plt.subplot(1, 2, 2)
plt.scatter(days_train, y_train.numpy().flatten() - train_predictions_ranfis)
plt.axhline(0, c='red', linestyle='--')
plt.title('Error')

plt.tight_layout()  
plt.show()
```

```python
ranfis.eval()
with torch.no_grad():
    val_predictions_ranfis = ranfis(X_val)[0].numpy().flatten()

plt.figure(figsize=(15, 4))

plt.subplot(1, 2, 1)
plt.scatter(days_val, y_val.numpy(), label='True Values', alpha=0.5)
plt.scatter(days_val, val_predictions_ranfis, label='Predictions')
plt.legend()
plt.title("Validation Data")

plt.subplot(1, 2, 2)
plt.scatter(days_val, y_val.numpy().flatten() - val_predictions_ranfis)
plt.axhline(0, c='red', linestyle='--')
plt.title('Validation Error')

plt.tight_layout()
plt.show()
```

```python
ranfis.eval()
with torch.no_grad():
    test_predictions_ranfis = ranfis(X_test)[0].numpy().flatten()

plt.figure(figsize=(15, 4))

plt.subplot(1, 2, 1)
plt.scatter(days_test, y_test.numpy(), label='True Values', alpha=0.5)
plt.scatter(days_test, test_predictions_ranfis, label='Predictions')
plt.legend()
plt.title("Test Data")

plt.subplot(1, 2, 2)
plt.scatter(days_test, y_test.numpy().flatten() - test_predictions_ranfis)
plt.axhline(0, c='red', linestyle='--')
plt.title('Test Error')

plt.tight_layout()
plt.show()
```

## Final Analysis

```python
rmse_rnn = np.sqrt(np.mean((y_test.ravel().numpy() - test_predictions_rnn)**2)).item()
rmse_ranfis = np.sqrt(np.mean((y_test.ravel().numpy() - test_predictions_ranfis)**2)).item()

print(f'RNN    | RMSE: {rmse_rnn:.5f} | Parameters: {params_rnn}')
print(f'RANFIS | RMSE: {rmse_ranfis:.5f} | Parameters: {params_ranfis}')
```

```python
days_pred = np.concatenate([days_val, days_test], axis=0) 
pred_rnn = np.concatenate([val_predictions_rnn, test_predictions_rnn], axis=0)
pred_ranfis = np.concatenate([val_predictions_ranfis, test_predictions_ranfis], axis=0)

plt.figure(figsize=(12, 8)) 

plt.subplot(2, 1, 1)  
plt.scatter(mean_days, mean_cases_day, label='Data', color='blue', s=15)
plt.axvline(days_train.max(), linestyle='--', color='black', label='Train Split')
plt.scatter(days_pred, pred_rnn, label='RNN Predictions', color='red', marker='s', s=15, alpha=0.7)
plt.legend()

plt.subplot(2, 1, 2)  
plt.scatter(mean_days, mean_cases_day, label='Data', color='blue', s=15)
plt.axvline(days_train.max(), linestyle='--', color='black', label='Train Split')
plt.scatter(days_pred, pred_ranfis, label='RANFIS Predictions', color='red', marker='s', s=15, alpha=0.7)
plt.legend()

plt.tight_layout()
plt.show()
```
