# Wavefunction Approximation 

```python
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
import pickle

from ANFISpy import ANFIS
```

Plotting.

```python
mfshape = 'sigmoid'
colors = ['blue', 'orange', 'red', 'green']  

for i, gridsize in enumerate([100, 200, 300, 400]):
    train_losses = np.load(f'data_{gridsize}/train_{mfshape}_{gridsize}.npy')
    val_losses = np.load(f'data_{gridsize}/val_{mfshape}_{gridsize}.npy')

    plt.plot(train_losses, color=colors[i], label=f'Train {gridsize}')
    plt.plot(val_losses, color=colors[i], linestyle='dashed', label=f'Validation {gridsize}')

plt.xlabel('Epoch', fontsize=14)
plt.ylabel('Loss (MSE)', fontsize=14)
plt.yscale('log')
plt.legend()
plt.title('Sigmoid', fontsize=16)
plt.show()
```

```python
mfshape = 'gaussian'
colors = ['blue', 'orange', 'red', 'green']  

for i, gridsize in enumerate([100, 200, 300, 400]):
    train_losses = np.load(f'data_{gridsize}/train_{mfshape}_{gridsize}.npy')
    val_losses = np.load(f'data_{gridsize}/val_{mfshape}_{gridsize}.npy')

    plt.plot(train_losses, color=colors[i], label=f'Train {gridsize}')
    plt.plot(val_losses, color=colors[i], linestyle='dashed', label=f'Validation {gridsize}')

plt.xlabel('Epoch', fontsize=14)
plt.ylabel('Loss (MSE)', fontsize=14)
plt.yscale('log')
plt.legend()
plt.title('Gaussian', fontsize=16)
plt.show()
```

```python
mfshape = 'bell'
colors = ['blue', 'orange', 'red', 'green']  

for i, gridsize in enumerate([100, 200, 300, 400]):
    train_losses = np.load(f'data_{gridsize}/train_{mfshape}_{gridsize}.npy')
    val_losses = np.load(f'data_{gridsize}/val_{mfshape}_{gridsize}.npy')

    plt.plot(train_losses, color=colors[i], label=f'Train {gridsize}')
    plt.plot(val_losses, color=colors[i], linestyle='dashed', label=f'Validation {gridsize}')

plt.xlabel('Epoch', fontsize=14)
plt.ylabel('Loss (MSE)', fontsize=14)
plt.yscale('log')
plt.legend()
plt.title('Generalized Bell', fontsize=16)
plt.show()
```

Defining metrics.

```python
gridsizes = [100, 200, 300, 400]

sigmoid_rmse = np.array([0.0379, 0.0235, 0.0185, 0.0157])
sigmoid_mae = np.array([0.0210, 0.0119, 0.0078, 0.0070])
sigmoid_r2 = np.array([0.9150, 0.9643, 0.9804, 0.9856])

bell_rmse = np.array([0.0259, 0.0197, 0.0171, 0.0116])
bell_mae = np.array([0.0145, 0.0110, 0.0086, 0.0068])
bell_r2 = np.array([0.9602, 0.9749, 0.9831, 0.9921])

gaussian_rmse = np.array([0.0478, 0.0196, 0.0269, 0.0215])
gaussian_mae = np.array([0.0283, 0.0126, 0.0129, 0.0126])
gaussian_r2 = np.array([0.8649, 0.9753, 0.9582, 0.9729])
```

```python
colors = ['blue', 'orange', 'red', 'green']  
    
plt.plot(gridsizes, sigmoid_rmse, marker='o', color=colors[0], label='Sigmoid')
plt.plot(gridsizes, gaussian_rmse, marker='o', color=colors[1], label='Gaussian')
plt.plot(gridsizes, bell_rmse, marker='o', color=colors[2], label='Generalized Bell')

plt.xlabel('Grid Size', fontsize=14)
plt.ylabel('Metric (RMSE)', fontsize=14)
plt.xticks(gridsizes)
plt.legend()
plt.title('RMSE', fontsize=16)
plt.show()
```

```python
colors = ['blue', 'orange', 'red', 'green']  
    
plt.plot(gridsizes, sigmoid_mae, marker='o', color=colors[0], label='Sigmoid')
plt.plot(gridsizes, gaussian_mae, marker='o', color=colors[1], label='Gaussian')
plt.plot(gridsizes, bell_mae, marker='o', color=colors[2], label='Generalized Bell')

plt.xlabel('Grid Size', fontsize=14)
plt.ylabel('Metric (MAE)', fontsize=14)
plt.xticks(gridsizes)
plt.legend()
plt.title('MAE', fontsize=16)
plt.show()
```

```python
colors = ['blue', 'orange', 'red', 'green']  
    
plt.plot(gridsizes, sigmoid_r2, marker='o', color=colors[0], label='Sigmoid')
plt.plot(gridsizes, gaussian_r2, marker='o', color=colors[1], label='Gaussian')
plt.plot(gridsizes, bell_r2, marker='o', color=colors[2], label='Generalized Bell')

plt.xlabel('Grid Size', fontsize=14)
plt.ylabel(r'Metric (R$^2$)', fontsize=14)
plt.xticks(gridsizes)
plt.legend()
plt.title(r'R$^2$', fontsize=16)
plt.show()
```
